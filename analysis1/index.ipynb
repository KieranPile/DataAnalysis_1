```python
numeric_columns = q1df.columns.difference(['Capital', 'State'])
q1df[numeric_columns] = q1df[numeric_columns].apply(pd.to_numeric, errors='coerce')
# Converts every row aside from the initial name to a numerical value 

q1df['Total Cost'] = q1df[numeric_columns].sum(axis=1)
# Calculates the sum of each numerical row

least_expensive = q1df.loc[q1df['Total Cost'].idxmin()]
# Finds the lowest result and saves it under a new variable 

print("The least expensive capital state to live in is:")
print(least_expensive[['Capital', 'State', 'Total Cost']]) 
# Displays the findings 

# This code can be altered to find other instances of data such as the most expensive by changing 'idxmin' to 'idxman' etc.
```

> It is shown that Dover is the least expensive state with a total cost of $1574.29.
